```python
#importing lib
import pandas as pd
import numpy as np
```

```python
#loading the data
data = pd.read_csv('D://Projects Data Analytics For Students//Projects Data Analytics For Students//Project 1//pupae.csv')
```

```python
#handling the missing values
data.fillna(method = 'ffill', inplace = True)
```

```python
#Convert ’CO2_treatment’ to a factor/categorical discrete variable. Inspect the levels of this factor variable.
```

```python
data["CO2_treatment"]=data["CO2_treatment"].astype(str) +" " +"ppm"
```

```python
data.head()
```

```python
#importing plotting lib
import matplotlib.pyplot as plt
```

```python
#Make a scatter plot of Frass vs. PupalWeight, with blue solid circles for a CO2 concentration of
#280ppm and red for 400ppm. Also add a legend.
```

```python
#sorting according to Co2 treatment
data.sort_values(by='CO2_treatment' , ignore_index = True , inplace = True)
```

```python
data.head(5)
```

```python
ppm1 = plt.scatter(data.Frass[0:39],data.PupalWeight[0:39])
#ppm280.set_label('280 ppm')
ppm2 = plt.scatter(data.Frass[39:],data.PupalWeight[39:],c='r')
#ppm2.set_label('400 ppm')
plt.xlabel('Frass')
plt.ylabel('Pupal Weight')
plt.title('Pupal Weight VS Frass')
plt.legend([ppm1,ppm2],['280 ppm','400 ppm'],title = 'CO2 treatment')
plt.show()
```

```python
#The problem with the above figure is that data for both temperature treatments is combined.
#Make two plots (either in a PDF, or two plots side by side), one with the ’ambient’ temperature
#treatment, one with ’elevated’.
```

```python
#sorting by Temperature treatment
data.sort_values(by='T_treatment' , ignore_index = True , inplace = True)
```

```python
data.head(5)
```

```python
#creating a new database of ambient temp treatment
data_aT = data[:37]
```

```python
data_aT.head(14)
```

```python
#sorting the ambient temp treatment database with CO2 treatment
data_aT.sort_values(by='CO2_treatment' , ignore_index = True , inplace = True)
```

```python
# creating a new database with elevated temp treatment
data_eT = data[37:]
```

```python
data_eT.head(25)
```

```python
#sorting the elevated temp treament database with CO2 treatment
data_eT.sort_values(by='CO2_treatment' , ignore_index = True , inplace = True)
```

```python
#plotting both the graph separatly
#plt.subplot(1,2,1)
ppm1 = plt.scatter(data_aT.Frass[0:14],data_aT.PupalWeight[0:14])
ppm2 = plt.scatter(data_aT.Frass[14:],data_aT.PupalWeight[14:],c='r')
plt.xlabel('Frass')
plt.ylabel('Pupal Weight')
plt.xlim(0.5,3)
plt.ylim(0.15,0.5)
plt.legend([ppm1,ppm2],['280 ppm','400 ppm'])
plt.title("For Ambient Temperature")

plt.show()

#plt.subplot(1,2,2)
ppm11 = plt.scatter(data_eT.Frass[0:25],data_eT.PupalWeight[0:25])
ppm22 = plt.scatter(data_eT.Frass[25:],data_eT.PupalWeight[25:],c='r')
plt.xlabel('Frass')
plt.ylabel('Pupal Weight')
plt.xlim(0.5,3)
plt.ylim(0.15,0.5)
plt.legend([ppm11,ppm22],['280 ppm','400 ppm'])
plt.title("For Elevated Temperature")

plt.show()
```

```python
#combining both the graph but using diffrent symbols for temperature treatment


#plt.subplot(1,2,1)
ppm1 = plt.scatter(data_aT.Frass[0:14],data_aT.PupalWeight[0:14],c='b')
ppm2 = plt.scatter(data_aT.Frass[14:],data_aT.PupalWeight[14:],c='r')
#plt.xlabel('Frass')
#plt.ylabel('Pupal Weight')
#plt.xlim(0.5,3)
#plt.ylim(0.15,0.5)
#plt.legend([ppm1,ppm2],['280 ppm','400 ppm'])
#plt.title("For Ambient Temperature")

#plt.show()

#plt.subplot(1,2,2)
ppm11 = plt.scatter(data_eT.Frass[0:25],data_eT.PupalWeight[0:25],c='w',marker = 's',
    edgecolors='b')
ppm22 = plt.scatter(data_eT.Frass[25:],data_eT.PupalWeight[25:],c='w',marker = 's',
    edgecolors='r')
plt.xlabel('Frass')
plt.ylabel('Pupal Weight')
plt.xlim(0.5,5)
plt.ylim(0.15,0.5)
legend1 = plt.legend([ppm1,ppm2],['for 280 ppm','for 400 ppm'],loc=4, title = 'CO2 treatment' )
plt.legend([ppm1,ppm22],['ambient','elevated'],loc=1, title = 'T treatment')
plt.gca().add_artist(legend1)
#plt.title("For Elevated Temperature")

plt.show()
```

```python
# all legends in one label
ppm1 = plt.scatter(data_aT.Frass[0:14],data_aT.PupalWeight[0:14],c='b')
ppm2 = plt.scatter(data_aT.Frass[14:],data_aT.PupalWeight[14:],c='r')

ppm11 = plt.scatter(data_eT.Frass[0:25],data_eT.PupalWeight[0:25],c='w',marker = 's',
    edgecolors='b')
ppm22 = plt.scatter(data_eT.Frass[25:],data_eT.PupalWeight[25:],c='w',marker = 's',
    edgecolors='r')
plt.xlabel('Frass')
plt.ylabel('Pupal Weight')
plt.xlim(0.5,5)
plt.ylim(0.15,0.5)
#legend1 = plt.legend([ppm1,ppm2],['for 280 ppm','for 400 ppm'],loc=4, title = 'CO2 treatment' )
plt.legend([ppm1,ppm2,ppm11,ppm22],['280:T_ambient','400:T_ambient','280:T_elevated','400:T_elevated'],loc=1, title = 'CO2:T',
           title_fontsize = 'x-large',shadow = True)
#plt.gca().add_artist(legend1)
#plt.title("For Elevated Temperature")

plt.show()
```
